```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
from rm3neng.simulators.basic import *

```

```python
ss, st, templates = generate_network_signal(4, [600, 1000], 40, 200000, 6)
print(templates.shape)
plt.figure()
plt.subplot(121)
plt.plot(ss[:2000])
plt.subplot(122)
plt.plot(templates)
plt.show()
```

```python
from scipy.signal import find_peaks

pk_locs = find_peaks(np.abs(ss), height=0.5, distance=30)[0]

spike_matrix = []
for p in pk_locs:
    spike_matrix.append(ss[p-30:p+30])
spike_matrix = np.asarray(spike_matrix)

print(spike_matrix.shape)
plt.figure()
plt.plot(spike_matrix.transpose(),'k')
plt.show()
```

```python
features = []

for s in range(spike_matrix.shape[0]):
    features.append([
        np.max(spike_matrix[s,:]),
        np.abs(np.min(spike_matrix[s,:])),
        np.ptp(spike_matrix[s,:]),
        np.sum(np.abs(spike_matrix[s,:])),
        ])
features = np.asarray(features)
print(features.shape)
```

```python
import seaborn as sns

sns.pairplot(pd.DataFrame(features))
```

```python
from itertools import combinations

comb = combinations(range(features.shape[1]),3)
fig = plt.figure(figsize=(20,20))
for i,c in enumerate(comb):
    ax = fig.add_subplot(2,2,i+1, projection='3d')
    ax.scatter(features[:,c[0]],features[:,c[1]],features[:,c[2]])
plt.show(fig)
```

```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

pca = PCA(n_components=4)
feat_pca = pca.fit_transform(StandardScaler().fit_transform(features))
print(feat_pca.shape)
```

```python
%matplotlib widget
fig = plt.figure()
ax = plt.subplot(121,projection='3d')
ax.scatter(feat_pca[:,0], feat_pca[:,1], feat_pca[:,2])
ax = plt.subplot(122,projection='3d')
ax.scatter(feat_pca[:,-1], feat_pca[:,-2], feat_pca[:,-3])
plt.show(fig)
```
